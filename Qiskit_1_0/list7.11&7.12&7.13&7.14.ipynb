```python
# 量子回路の初期化
circuit = QuantumCircuit(2,2)

# 量子回路の組み立て
circuit.h(0) # アダマールゲート行列を適用
circuit.x(0) # CNOTを適用
circuit.h(1) # アダマールゲート行列を適用

# 測定
circuit.measure([0,1],[0,1])

# 量子回路を可視化
circuit.draw(output="mpl")
```

```python
# 量子回路の初期化
circuit = QuantumCircuit(2,2)

# 量子回路の組み立て
circuit.h(0) # アダマールゲート行列を適用
circuit.x(0) # CNOTを適用
circuit.h(1) # アダマールゲート行列を適用
circuit.barrier()

# 測定
circuit.measure([0,1],[0,1])

# 量子回路を可視化
circuit.draw(output="mpl")
```

```python
# barrierを表示しない
circuit.draw(output="mpl", plot_barriers=False)
```
