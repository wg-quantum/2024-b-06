```python
# Random Walk 1D <Classical Random Walk in 1-Dimention>

#--- import
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from matplotlib import rcParams
from matplotlib.font_manager import FontProperties

#--- Font settings for linux(OpenSuse Tumbleweed)
font_path = "/usr/share/fonts/truetype/NotoSansCJKjp-Regular.otf"
font_prop = FontProperties(fname=font_path)

#--- Parameters
num_steps  = 1024    # Number of steps in one random walk
num_trials = 1024    # Number of times to repeat the random walk
total_counts = Counter()  # Count the overall frequency

#--- Perform multiple random walks
for _ in range(num_trials):
    position = [0]  # Initial position is 0 for each trial
    for _ in range(num_steps):
        step = np.random.choice([-1, 1])  # -1 moves left, 1 moves right
        position.append(position[-1] + step)
    # Add the results of each trial to the overall count
    total_counts.update(position)

#--- Calculate average frequency
positions = list(total_counts.keys())
frequencies = [total_counts[pos] / num_trials for pos in positions]

#--- Plot the results as a histogram
plt.bar(positions, frequencies, color='b')
plt.title(f'Average Frequency of 1D Random Walks ({num_steps} steps : {num_trials} trials)', fontproperties=font_prop)
plt.xlabel('Position', fontproperties=font_prop)
plt.ylabel('Average Frequency', fontproperties=font_prop)
plt.grid(True)

#--- Save the plot as a PNG file
plt.savefig('images/rw/random-walk-histogram.png')

#--- Show the plot
plt.show()
```
